## Libraries Used:

We'll be using two powerful libraries:

- **Matplotlib:** A versatile plotting library for creating static, animated, and interactive visualizations in Python.
- **Seaborn:** A statistical data visualization library based on Matplotlib. Seaborn provides a high-level interface for drawing attractive and informative statistical graphics.

---

### Plot 1: Number of Planets Discovered by Method

This plot shows the distribution of discovered planets based on the method used.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the "Planets" dataset from Seaborn
planets_data = sns.load_dataset('planets')

plt.figure(figsize=(10, 5))
sns.countplot(x='method', data=planets_data)
plt.title('Number of Planets Discovered by Method')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Plot 2: Distribution of Planetary Masses

Usage of histogram displays the spread of masses among the discovered exoplanets.

```python
plt.figure(figsize=(10, 5))
sns.histplot(x='mass', data=planets_data, bins=20, kde=True)
plt.title('Distribution of Planetary Masses')
plt.show()

```

### Plot 3: Orbital Period vs Planetary Mass

Lets try scatter plot to explore the relationship between the orbital period and planetary mass.

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='orbital_period', y='mass', data=planets_data)
plt.title('Orbital Period vs Planetary Mass')
plt.show()
```

### Plot 4: Distribution of Planetary Masses by Discovery Method

The violin plot illustrates the distribution of planetary masses, categorized by discovery method.

```python
plt.figure(figsize=(10, 5))
sns.violinplot(x='method', y='mass', data=planets_data)
plt.title('Distribution of Planetary Masses by Method')
plt.xticks(rotation=45, ha='right')
plt.show()

```

### Plot 5: Distribution of Planetary Distances

This boxplot displays the distribution of distances from Earth for the discovered exoplanets.



```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='distance', data=planets_data)
plt.title('Distribution of Planetary Distances')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Plot 6: Distribution of Planetary Distances by Discovery Method

The strip plot reveals the distribution of planetary distances categorized by the discovery method.

```python
plt.figure(figsize=(10, 5))
sns.stripplot(x='method', y='distance', data=planets_data, dodge=True, alpha=0.7)
plt.title('Distribution of Planetary Distances by Method')
plt.xticks(rotation=45, ha='right')
plt.show()
```
